# Feature Engineering

Creating new features like -
1. Bill-per-GB: This column calculates the ratio of total usage (in GB) to monthly bill. It provides insight into how much data usage a customer gets per unit cost of their monthly bill.
2. Usage_Bill_Interaction:
This column represents the interaction between total usage (in GB) and the monthly bill. It's the product of total usage and monthly bill. This interaction feature captures potential relationships between usage and billing.
3. GBpermonth:
This column calculates the average data usage per month by dividing total usage (in GB) by the subscription length (in months). It provides an understanding of the customer's average data consumption rate.
4. SubscriptionValue:
This column calculates the value of the subscription in terms of data usage. It's the ratio of total usage (in GB) to the product of subscription length (in months) and monthly bill. It indicates how much data usage a customer gets relative to their subscription cost.
5. Monthly_Bill_Category:
This categorical column categorizes customers based on their monthly bill. It bins the values of the Monthly_Bill column into predefined ranges using the pd.cut function. The categories are 'Low Bill', 'Medium Bill', and 'High Bill', indicating different levels of monthly billing.
6. Long_Term_Customer:
This binary column indicates whether a customer is a long-term subscriber or not. It is derived from the Subscription_Length_Months column. If the subscription length is greater than a predefined threshold (referred to as long_term_threshold), the value is set to 1; otherwise, it's set to 0. This helps identify customers who have subscribed for a long duration.


```python
df_preprocessed['Bill-per-GB'] = df_preprocessed['Total_Usage_GB']/df_preprocessed['Monthly_Bill']
```

```python
df_preprocessed['Usage_Bill_Interaction'] = df_preprocessed['Total_Usage_GB'] * df_preprocessed['Monthly_Bill']
```

```python
df_preprocessed
```

```python
df_preprocessed['GBpermonth'] = df_preprocessed['Total_Usage_GB']/df_preprocessed['Subscription_Length_Months']
```

```python
df_preprocessed['SubscriptionValue'] = df_preprocessed['Total_Usage_GB']/(df_preprocessed['Subscription_Length_Months']*df_preprocessed['Monthly_Bill'])
```

```python
bill_ranges = [0, 50, 100, float('inf')]
category_labels = ['Low Bill', 'Medium Bill', 'High Bill']

df_preprocessed['Monthly_Bill_Category'] = pd.cut(df_preprocessed['Monthly_Bill'], bins=bill_ranges, labels=category_labels, right=False)
```

```python
df_preprocessed['Monthly_Bill_Category'] = df_preprocessed['Monthly_Bill_Category'].cat.codes
```

```python
long_term_threshold = 12

df_preprocessed['Long_Term_Customer'] = df_preprocessed['Subscription_Length_Months'].apply(lambda x: 1 if x > long_term_threshold else 0)
```

# Model Training and Evaluation

```python
import tensorflow as tf
```

```python
X = df_preprocessed.drop('Churn', axis=1)
y = df_preprocessed['Churn']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.50, random_state=42)
```

```python
print("Train size : ", X_train.shape[0])
print("Validation size : ", X_val.shape[0])
print("Test size : ", X_test.shape[0])
```

# Neural Networks

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.legacy.Adam(learning_rate=0.001, decay=1e-5),
              metrics=['accuracy'])

early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)

# Train the model
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=32,
                    validation_data=(X_val, y_val),
                    callbacks=[early_stopping],
                    verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
```
